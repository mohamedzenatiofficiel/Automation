```python
# <div class="new_info_next">
#      <input type="hidden" value="133" id="new_id" class="new_id">
#      <input type="hidden" value="0" id="default_pe" class="default_pe">
# </div>
title=[]
link=[]
degres=[]
price = []

temp = ["text--color-charcoal", "vote-temp--warm", "vote-temp--cold", "vote-temp--colder", "vote-temp--hot"]
reduc=["l","1"]

info = soup.find_all("div", {"class": "threadGrid thread-clickRoot"})

for infos in info:
    titre = infos.find_all("a", {"class": "cept-tt thread-link linkPlain thread-title--list js-thread-title"})
    for inf in titre:
        title.append(inf.get('title'))
        link.append(inf.get('href'))
for degre in info:
    for deg in temp:
        temperature = degre.find_all("span", {"class": f"cept-vote-temp vote-temp {deg}"})
        for t in temperature:
            degres.append(t.text.strip())
            

for prix in info:
    for red in reduc:
        cost = prix.find_all("span", {"class": f"thread-price text--b cept-tp size--all-{red} size--fromW3-xl"})
        free = prix.find_all("span", {"class": "overflow--fade"})
        if not cost:
            for money in free:
                price.append(money.text.strip())
        else:
            for n in cost:
                price.append(n.text.strip())
      
        
                   
df_link=pd.DataFrame(link)
df_link.rename(columns={0: "lien"},inplace=True)        
df_link = df_link.reset_index(drop=True)   

singles = [singularize(plural) for plural in title]
df_titre=pd.DataFrame(singles)    
df_titre.rename(columns={0: "titre"},inplace=True)
df_titre = df_titre.reset_index(drop=True)

df_temperature = pd.DataFrame(degres)
df_temperature.rename(columns={0: "temperature"},inplace=True) 
df_temperature = df_temperature.reset_index(drop=True)



df_price = pd.DataFrame(price)
df_price.rename(columns={0: "prix"},inplace=True)
df_price = df_price.iloc[:-2:2]
df_price = df_price.reset_index(drop=True)



df_final = pd.concat([df_titre, df_temperature, df_price, df_link],axis=1)
df_final
```

```python
df_final['titre'] = df_final['titre'].str.lower()
df_final['titre'] = df_final['titre'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df_final['prix'] = df_final['prix'].str.lower()
df_final
```

```python
df_final['prix'] = df_final['prix'].replace('', 'none')
df_final['prix'] = df_final['prix'].fillna(0)
df_final['temperature'] = df_final['temperature'].fillna(0)
df_final['prix'] = df_final['prix'].astype(str)
df_final
```

```python
df_final = df_final.to_dict('records')
df_final
```
