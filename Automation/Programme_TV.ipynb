```python
import pandas as pd
from datetime import datetime, timedelta, date
import json
import requests
import xmltodict
```

```python
OUTPUT_FILE_LOCATION = 'C:/Users/zmoha/OneDrive - Efrei/Bureau/Automation/Programme.csv'
```

```python
def date(i):
    dates = datetime.strftime((datetime.now() + timedelta(days=i)), '%Y-%m-%d')
    return dates
```

```python
url = "https://xmltv.ch/xmltv/xmltv-tnt.xml"
response = requests.get(url)
data = xmltodict.parse(response.content)
dictionnary = data['tv']['programme']

df = pd.DataFrame(dictionnary)
df
```

```python
df = df.replace({'C192.api.telerama.fr' : 'TF1',
    'C4.api.telerama.fr' : 'France 2',
    'C80.api.telerama.fr' : 'France 3',
    'C34.api.telerama.fr' : 'Canal +',
    'C111.api.telerama.fr' : 'Arte',
    'C118.api.telerama.fr' : 'M6',
    'C445.api.telerama.fr' : 'C8',
    'C119.api.telerama.fr' : 'W9',
    'C195.api.telerama.fr' : 'TMC',
    'C446.api.telerama.fr' : 'TFX',
    'C444.api.telerama.fr' : 'NRJ 12',
    'C78.api.telerama.fr' : 'France 4',
    'C234.api.telerama.fr' : 'La Chaîne parlementaire',
    'C481.api.telerama.fr' : 'BFMTV',
    'C226.api.telerama.fr' : 'CNEWS',
    'C458.api.telerama.fr' : 'CSTAR',
    'C482.api.telerama.fr' : 'Gulli',
    'C1404.api.telerama.fr' : 'TF1 Séries Films',
    'C1401.api.telerama.fr' : "L'Equipe",
    'C1403.api.telerama.fr' : '6ter',
    'C1402.api.telerama.fr' : 'RMC Story',
    'C1400.api.telerama.fr' : 'RMC Découverte',
    'C1399.api.telerama.fr' : 'Chérie 25',
    'C112.api.telerama.fr' : 'LCI',
    'C2111.api.telerama.fr' : 'Franceinfo',
    'C47.api.telerama.fr' : 'France 5'})


df[["Date_start", "Useless1"]] = df["@start"].str.split(pat=" ", expand=True)
df[["Date_stop", "Useless2"]] = df["@stop"].str.split(pat=" ", expand=True)
df = df.drop(["Useless1", "Useless2"], axis=1)
df['Date_start'] = df['Date_start'].apply(lambda x: datetime.strptime(x, '%Y%m%d%H%M%S').strftime('%Y-%m-%d %H:%M:%S'))
df['Date_stop'] = df['Date_stop'].apply(lambda x: datetime.strptime(x, '%Y%m%d%H%M%S').strftime('%Y-%m-%d %H:%M:%S'))

df = df.rename(columns={
                        "@channel": "Channel",
                        "title": "Programme",
                        "category": "Categorie"
                       })

df = df.drop(['sub-title', 
              'desc', 
              'date',  
              'icon', 
              'episode-num', 
              'audio', 
              'rating', 
              'previously-shown', 
              'premiere', 'new', 
              'star-rating', 
              'subtitles'], axis=1)

df
```

```python
df_categorie = pd.json_normalize(df['Categorie'])
df_categorie = df_categorie.rename(columns={"#text": "Type programme", "@lang": "Langue"})
df_categorie
```

```python
df_time = pd.json_normalize(df['length'])
df_time = df_time.rename(columns={"#text": "Duree programme", "@units": "unite"})
df_time
```

```python
df_programme = pd.concat([df_time, df_categorie, df], axis=1)
df_programme[['Date_start','Start']] = df.Date_start.str.split(" ",expand=True,)
df_programme[['Date_stop','Finish']] = df.Date_stop.str.split(" ",expand=True,)

df_programme = df_programme.drop(['Categorie', 'length'], axis=1)

df_programme = df_programme[["Channel", "Date_start", "Date_stop", "Programme", "Type programme", "Start", "Finish", "Duree programme", "unite", "Langue"]]

df_programme
```

```python
# df_campaign_report_conversion = df_campaign_report_conversion[(df_campaign_report_conversion["Conversions name"]=='Macro - Leads Call Back') | (df_campaign_report_conversion["Conversions name"]=='Macro - Leads Funnel')]
filtered_df = df_programme[(df_programme["Date_start"]==date(0)) | (df_programme["Date_start"]==date(1)) | (df_programme["Date_start"]==date(2))]

# filtered_df = df_programme[df_programme.Date_start.eq(yesterday_date)]
filtered_df = filtered_df.rename(columns={"Date_start": "Date start",
                                          "Date_stop": "Date stop"
                       })
filtered_df.reset_index(drop=True, inplace=True)

filtered_df
```

```python
filtered_df.to_csv(OUTPUT_FILE_LOCATION, encoding='utf-8', index=False, sep='|')
filtered_df
```
